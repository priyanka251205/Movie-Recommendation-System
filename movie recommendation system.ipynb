```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english')
vectors = cv.fit_transform(df['tags'])
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
df[df['title'].str.contains("love")]
```

```python
def recommend(movie_title, top_n=5):
    movie_title = movie_title.lower()
    if movie_title not in df['title'].values:
        return []
    idx = df[df['title'] == movie_title].index[0]
    scores = list(enumerate(similarity[idx]))
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    return [df.iloc[i[0]]['title'] for i in sorted_scores]
```

```python
recommend("lagaan: once upon a time in india")
```

```python
recommend("hum tumhare hain sanam")
```
